```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Internal': [450, 500, 520, 560, 600, 620, 650,700,720,700,800,830,870,900,950],
    'External': [8200, 9100, 9400, 9900, 10500, 11000, 11500,12100,12500,13200,13900,14400,15100,15800,16500]
}

df = pd.DataFrame(data)
```

```python
#define x nd y
x= df[['Internal']]
y= df['External']
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
```

```python
y_pred=model.predict(x)
```

```python
print('coefficient',model.coef_)
print('intercept',model.intercept_)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
import numpy as np
mae= mean_absolute_error(y,y_pred)
mse= mean_squared_error(y,y_pred)
rmse=np.sqrt(mse)

print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x,y_pred,color='red')
plt.xlabel("internal Marks")
plt.ylabel("external marks")
plt.title("linear regression")
plt.show
```
